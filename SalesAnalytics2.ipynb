```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")

import seaborn as sns
```

```python
path =r"/content/Sales.csv"
df = pd.read_csv(path, encoding = "unicode-escape")
```




Data Imputation




```python
mean_val = df["Amount"].mean()
mean_val

df["Amount"]= df["Amount"].replace(np.NaN, mean_val)
```

```python
df.isnull().sum()
```

Changing DataType

```python
df["Amount"] = df["Amount"].astype("int64")
```

```python
df.info()
```

Feature selection

```python
df.drop(columns =["Status","unnamed1" ], axis =1, inplace=True)
```

```python
df.columns
```

Data Visualization

```python
ax=sns.countplot(data=df, x= "Gender", hue ="Gender")


for i in ax.containers:
  ax.bar_label(i)
```

```python

DFsum=df.groupby(["Gender"], as_index =False)["Amount"].sum()    .sort_values(by=["Amount"],inplace =False, ascending =False)
DFsum
```

```python
sns.barplot(data =DFsum, x = "Gender", y = "Amount", hue ="Gender")

# The number of females are a lot more, thus females contirbute to the bulk of the revenue
```

```python
DFavg = df.groupby(["Gender"], as_index =False)["Amount"].mean() .sort_values(by =["Amount"], ascending =False, inplace =False)
DFavg
```

```python
ax =sns.barplot(data = DFavg, x ="Gender", y ="Amount", hue ="Gender")

for i in ax.containers:
  ax.bar_label(i)


  # Females on an average spend a little more than men
```

```python
ax= sns.countplot(data =df , x ="Age Group", hue ="Gender")

for i in ax.containers:
  ax.bar_label(i)

# Women from the age range of 26-35  and 36-45 have the highest number of users
```

```python
DFageMean = df.groupby(["Age Group"], as_index=False)["Amount"].mean()    .sort_values(by=["Amount"], ascending=False, inplace=False)
DFageMean
```

```python
ax =sns.barplot(data = DFageMean, x ="Age Group", y = "Amount", hue = "Age Group")

for i in ax.containers:
  ax.bar_label(i)


#the age group 51-55 sepnds the most
```

```python
df.corr().unstack().sort_values(ascending=False)

# There are no major trends in the data
```

```python
dfStateSum = df.groupby(["State"], as_index =False)["Amount"].sum() .sort_values(by =["Amount"], ascending =False) .head(10)
dfStateSum
```

```python
sns.barplot(data =dfStateSum, x ="Amount", y="State", hue ="Amount")

#Top 10 states where the customers living have spent the most
```

```python
dfOrderSum = df.groupby(["State"], as_index =False)["Orders"].sum() .sort_values(by =["Orders"], ascending =False) .head(10)
dfOrderSum
```

```python
sns.barplot(data = dfOrderSum, y ="State", x ="Orders", hue ="Orders")

# From the above to visualization we can confirm that both the states have similar purchasing power
```

Marital Status

```python
ax = sns.barplot(data =df , x ="Marital_Status", y ="Amount" , hue ="Gender")


for i in ax.containers:
  ax.bar_label(i)

#unmarried(0) women have spent the most
```

```python
dfSUMsales = df.groupby(["Occupation"], as_index = False)["Amount"].sum().sort_values(by =["Amount"], ascending=False)
dfSUMsales
```

```python
plt.figure(figsize=(20, 6))

sns.countplot(data =dfSUMsales , x ="Occupation", hue= "Amount")
plt.show()

#People from IT, healthcare and aviation sectors spend the most.
```

```python
dfSUMprod = df.groupby(["Product_Category"], as_index = False)["Amount"].sum().sort_values(by =["Amount"], ascending=False)
dfSUMprod
```

```python
ax = sns.barplot(data = dfSUMprod, x = "Amount", y = "Product_Category", hue ="Amount", palette='dark:blue')

```

```python
dfSUMprodid = df.groupby(["Product_ID"], as_index = False)["Orders"].sum().sort_values(by =["Orders"], ascending=False).head(10)
dfSUMprodid
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(data = dfSUMprodid, x = "Product_ID", y = "Orders",  hue ="Orders")

```

We can conclude that married women from the age group 25-35 from the state of Uttar Pradesh and Maharashtra working in IT and Healthcare sector are more likely to buy products like food items and clothing apparels.
